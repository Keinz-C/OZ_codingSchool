[문제 2: 입금]

가지고 있는 돈을 입금합니다.

입금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 입금 거래를 영수증 리스트에 기록하세요.

- 입금 금액은 input()을 이용해 받습니다.
- 영수증은 list()로 선언합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.


````python
# 코드가 실행되며 초기화되는 것을 막기 위해서 영수증은 따로 빼놓음.
receipt = []
````

````python
# 함수를 설정해서 while문이 작동할 때 account에 돈이 쌓이고 잔액을 출력할 수 있도록 설정
def deposit(account, money):
    account += money
    print(f'입금되었습니다. 현재 남아있는 잔액은 {account}원 입니다.')
    return account  # input된 입금금액을 지속적으로 쌓일 수 있도록 반환 작성

# while문은 input값이 True를 반환하면 지속적으로 입금을 이어가고, Error나 0을 입력하여 입금을 강제 중지, 중단 시킬 수 있도록 설정.
while True:
    try:
        money = int(input('입금할 금액은 정수로만 작성해주세요 (종료하려면 0을 입력하세요):'))
        if money == 0:
            print('입금을 종료합니다.')
            break
        account = deposit(account, money)
    except ValueError:
        print('잘못된 입력입니다. 정수를 입력해주세요.')
````

[문제 3: 출금]

사용자가 돈을 출금합니다. 출금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 출금 거래를 영수증 리스트에 기록하세요. 

- 출금 금액은 input()을 이용해 받습니다.
- 현재 잔액보다 출금 요청 금액이 많을 경우 현재 잔액 만큼만 출금되도록 합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

````python
def withdraw(account, money2):
  if money2 <= account:
    account -= money2
    print('출금{0}원 : 현재 잔액 {1}'.format(money2, account))
    return account
  else:
    print('출금액이 잔액보다 많습니다.')
  
while True:
  try:
    money2 = int(input('출금하실 금액을 숫자로만 작성해주세요.(종료를 원하시면 0을 입력해주세요.)'))
    if money2 == 0:
      print('출금을 종료합니다.')
      break
    account = withdraw(account, money2)
  except ValueError:
    print('잘못 입력하였습니다. 시스템을 종료합니다.')


# def withdraw(balance, money):
#   if balance >= money:
#     print("출금이 완료되었습니다. 잔액은{0}원입니다.".format(account))
#     return balance - money
#   else:
#     print("출금이 완료되지 않았습니다. 잔액은 {0}원입니다.".format(account))
#     return balance

# balance = 0
# balance = deposit(balance,1000)
# balance = withdraw(balance, 2000)
# balance = withdraw(balance, 500)
# print(balance)
````
