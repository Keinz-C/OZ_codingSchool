### 변수란? ( 기본중의 기본이면서 근본이면서 코어면서... 암튼 그저 **GOAT**)

1. 하나의 변수에 하나의 값을 넣을때
*   변수이름 = 값

2. 두개의 변수에 하나의 값을 각각 넣을때
*   변수이름1, 변수이름2 = 값1, 값2

```python
# 하나의 값을 변수에 넣을 때
num = 1
# print(num)

# 두 개의 값을 변수에 넣을 때
num_1, num_2 = 1, 2
print(num_1)
print(num_2)

print(num_1, num_2)
```

### 변수에 값을 넣는 것 => 할당

sum = 1 + 2 => 1 + 2의 계산된 결과를 sum이라는 변수에 할당한다.  
sum 변수명이고 "=" 같다는 의미가 아니라 할당(assignment)이라고 한다.  
"=" : 할당 연산자  
"+" : 산술 연산자  
"+=" : 산술 후 바로 할당하는 문법  
/=, -=, *=도 모두 가능한다.

```python
# 변수를 사용하면서 사용 가능한 문법

sum = 1 + 2
sum = sum + 3
sum = sum + 4
sum = sum + 5

print(sum)

sum2 = 1 + 2
sum2 += 3
sum2 += 4 # sum = sum + 4와 문법적으로 같다.
sum2 += 5

print(sum2)
```

```python
# 주의 !! 선언되지 않은 변수를 사용할 경우 에러 발생

sum3 = 0
sum3 += 1

sum3
```

###변수를 만드는 규칙  
영문 문자와 숫자를 사용할 수 있다.  
대소문자를 구분한다.  
첫 글자는 대문자로 쓰는걸 지향한다.  
문자부터 시작해야 하며 숫자부터 시작하면 안 된다.  
_(밑줄 문자)로 시작할 수 있다.  
특수 문자(+, -, *, /, @, &, %등)는 사용할 수 없다.  
파이썬의 키워드(if, for, while, and, or 등)는 사용할 수 없다.

```python
# 대소문자를 구분하는 변수
BIG = "크다"
big = "작다"

print(BIG, big)
```
