```python
이름 = "햄스터"
나이 = 2
키 = 20.2
소속 = "드정글리안"

print(type(이름))
print(type(나이))
print(type(키))
print(type(소속))
```

```python
print(float(5))
print(int(5.0))
str_1 = "5"
str_2 = "5.0"
print(type(str_1), type(str_2))
```

시퀀스란? 요소들이 연속적으로 이어진 자료형이며  
값 하나하나를 요소라고 한다.

```python
# 대괄호 안 0은 index라고 한다.
# index : 0부터 시작한다.
print(소속)
print(소속[0])
print(소속[1])
print(소속[2])
print(소속[3])
print(소속[4])
```

```python
"드" in 소속
"햄" in 소속
```

```python
# 문자열 즉 시퀀스끼리 합치기

"13" + "금요일"
```

```python
# 시퀀스 자료형은 복사가 가능하다.

옛날게임 = "펭귄 브라더스" * 20
print(옛날게임)

len(옛날게임)
```

```python
생년월일 = "2000.12.25"

# 2000 12 25로 출력해보자

년 = 생년월일[0:4]
월 = 생년월일[5:7]
일 = 생년월일[8:11]

print(f'{년} {월} {일}') # 처음엔 년월일만 줬지만 다음엔 f스트링도 써보는걸로
```

```python
숫자 = "123456789"
# 전체 출력
print(숫자[::])

# 전체 출력 마지막은 요소이기 때문에 순서에 영향을 줌
print(숫자[::-1])

# 첫 번째 인덱스 기준으로 2칸씩 띄어쓰기 출력
print(숫자[::2])

print(숫자[4:1:-1])
print(숫자[4:])
```

```python
퀴즈 = "스님이 공중에 뜬다를 4글자로 말하면? 어중이떠중이"

# 어중이떠중이라는 정답만 출력하자
print(퀴즈[-7:])
```

```python
라라랜드 = "사람들은 다른 사람들의 열정에 끌리게 되어있어 자신이 잊은 걸 상기시켜 주니까"

# 라라랜드의 type을 알아보자!
type(라라랜드)
print(type(라라랜드))

# 라라랜드의 글자 수는?
len(라라랜드)
print(f'{len(라라랜드)}자 입니다.')

# 라라랜드에서 사람이라는 단어는 몇 번 나와?
라라랜드.count("사람")
print(f'{라라랜드.count("사람")}번 입니다.')

# 라라랜드에서 정열이라는 단어가 있어?
"정열" in 라라랜드

# 마음에 드는 문구끼리 떼어내서 붙이고 싶네... 해보자!
like_1 = 라라랜드[0:4]
like_2 = 라라랜드[13:25]
print(f'{like_1} {like_2}')
```
