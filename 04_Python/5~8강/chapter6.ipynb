### 형변환 : 말 그대로 형태를 변형한다

int, float  
int -> float 예) 5 -> 5.0 float(5)

```python
print(float(5))
print(int(5.0))
```

```python
나이 = 10
type(나이)
print(dir(나이))
```

```python
print(나이.bit_length()) # bit_length : 비트의 길이를 구해주는 내장 메서드
print(bin(나이)) # bin : 변수의 비트를 구해주는 함수
```
