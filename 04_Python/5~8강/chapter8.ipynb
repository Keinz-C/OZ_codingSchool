```python
# 본격적인 논리연산자
# True : 1 / false : 0

True and True # True는 1이고 and는 *로 인식되어 / 1 * 1 = 1
True and False # 1 * 0 = 0
False and True # 0 * 1 = 0
False and False # 0 * 0 = 0
```

```python
# 본격적인 논리연산자
# True : 1 / false : 0

True or True # True는 1이고 or는 +로 인식되어 / 1 + 1 = 1
True or False # 1 + 0 = 1
False or True # 0 + 1 = 1
False or False # 0 * 0 = 0
```

```python
# 점점 복잡해지는 조건들!!

10 == 12 or 6 != 2 # True / 10과 12는 같지 않아 False출력, 6과 2는 같지 않다는 조건이 성립되어 True출력으로 False or True는 +개념으로 True출력
2 > 5 and 7 == 7 # False / 2는 5보다 크다는 False출력, 7과 7은 같다는 True출력으로 Fasle * True = False출력
not 14 <= 15 # False / 14는 15보다 작거나 같다는 True 출력, 하지만 앞의 not 논리연산으로 False출력
not 7 is 7.0 # True / 7과 7은 같다는 정수 7과 실수 7.0은 객체가 같지 않아 false, 하지만 앞의 not 논리연산으로 True 출력 / 정수와 실수는 배당되는 id 값의 차이도 있음.

print(id(7))
print(id(7.0))
```
