###사칙연산(헤더는 자연스럽게 한 줄 띄워진대요!)
더하기 : +

빼기 : -

곱하기 : *

나누기 :
*   / : 우리가 알고 있는 나누기와 동일한 기능
*   // : 몫을 반환 (예 : 5/2 = 2)
*   % : 나머지를 반환 (예 5/2 = 1)


거듭제곱 : **

```python
#더하기를 실행해보자!
print(3 + 2)
```

```python
#빼기를 실행해보자!
print(3-2)
```

```python
#곱하기를 실행해보자!
print(3*2)
```

```python
3*2
```

```python
# /를 활용한 나누기를 실행해보자!
print(7/2)

# //
print(7//2)

# %
print(7%2)
```

내장함수 : divmod

>는 인용할때 쓰는 마크다운 표기야!
> 내장함수 : 자주 사용하는 기능을 미리 만들어 놓은 것을 내장함수라고 함!

**divmod** : 몫과 나머지를 동시에 구해줌

```python
print(divmod( 3, 2 ))
print(divmod( 7, 4 ))
print(divmod( 13, 4 ))
print(divmod( 29, 13 ))
print(divmod( 40, 27 ))

print(( 7//4, 7%4 ))
print( 7//4, ",", 7%4 )
```

###괄호

수학에서 사용되는 괄호와 동일한 기능을 가진다.

**()** : 괄호안에 있는 것을 우선적으로 계산한다.

사칙연산 순서 : 괄호 -> 곱셉과 나눗셈 -> 덧셈과 뺄셈

```python
3 + 2 * 4
```

```python
( 3 + 2 )*4
```
