### while문 목차
1. while문 만들어보기
2. while문 코드의 흐름 보기
3. while문 조건식 증감 감소
4. while문을 조심스럽게 사용해야하는 이윺
5. while문이 for보다 강점인 부분
6. continue, break
7. while문에도 else가 있다!
8. 붕어빵 인터네셔널은 내가 살린다!
9. 예제로 while문 익혀보기

```python
# while문 만들어보기

i = 0 # 초기식
while i < 10: # while 조건식
    print('점점 성장하는 모습!') # 반복할 코드
    i += 1 # while문은 1이 증가하는 이 코드가 들어가야한다. / 이는 변환식이다.
```

### while문의 작동 방식

* 초기식 -> 조건식 -> 반복할 코드 및 변환식을 반복함.

```python
# while문 조건식 증감, 감소

# 기본적인 증가 유형

i = 1
while i < 10:
    print('이정도로 성장할줄이야...!', i)
    i += 1


# 기본적인 감소 유형

i = 10
while i > 0:
    print('이정도로 감소할줄이야', i)
    i -= 1
```

```python
# while문을 조심스럽게 사용해야하는 이유
# 조건이 잘못 설정되면 무한 루프에 빠져 과도한 리소스 소모
# 하지만, 잘 사용하면 for문보다 효율적

# while True:
#     print(1)
```

```python
# while문이 for보다 강점인 부분?

import random

# random.randint(1,6) # randint는 random하게 int(정수)를 (값, 값) 사이에서 추출함

i = 0
while i != 4:
    i = random.randint(1,6)
    print(i)
```

```python
# while문에도 else가 있다.

양파 = 10
while 양파 > 10:
    print("양파", 양파, "개 남았습니다.")
    if 양파 ==0:
        break
        양파 -= 1
else:
    print("양파 다 팔렸습니다.")
```

```python
# 반복문에서 흐름을 제어하는 continue와 break

i = 0
while True:
    print(i)
    i += 1
    if i == 10:
        break


for i in range(100):
    print(i)
    i += 1
    if i == 10:
        break
```

```python
# continue

i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i)


# for i in range(10):
#     if i % 2 == 0:
#         continue
#     print(i)
```

```python
# 예제로 while문 익혀보기

# 콜라츠라는 사람은 주어진 수가 1이 될때까지 다음 작업을 반복하면, 모든 수를 1로 만들 수 있다고 추측했다
# 작업
# 1-1. 입력된 수가 짝수라면 2로 나눈다.
# 1-2. 입력된 수가 홀수라면 3을 곱하고 1을 더한다.
# 3. 위 결과로 나온 수에 같은 작업을 1이 될 때까지 반복한다.
# 예를 들어, 입력된 수가 6이라면, 6->3->10->5->16->8->4->2->1

# num변수에 있는 수를 1로 만드는데 몇번이나 작업을 반복해야하는지 출력한다.
num = 7


while i in num:
    if i % 2 == 0:
        i * 3 + 1
    if i * 2 != 0:
        i / 2
        continue
        print(1)

# 문제점 1. i를 써야 된다 생각하고 num변수를 무조건 i에 대입하려했던 점
# 문제점 2. 초기화를 위해 정답 변수 0으로 초기화 하지 않은 점
# 문제점 3. else문을 사용하여 홀수 일시 제외 짝수는 바로 다음 조건으로 넘어갈 수 있는 점
# 문제점 4. 문제점 3과 이어 2개의 조건을 보며 실행하기 때문에 continue를 작성한다면 다시 처음 조건으로 돌아가 무한 루프가 된다는 점.
```

```python
num = 7

정답 = 0

while num != 1:
    if num % 2 == 0:
        num = num / 2
    else:
         num = num * 3 + 1
    정답 += 1

print(정답)
```
