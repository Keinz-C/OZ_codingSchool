###1. if문이란?
* 특정 조건일때 코드를 실행하는 문법

```
if 파이썬을 정복한다면:
    개발자로 취직할 수 있다.
```

###2. if문 문법과 주의사항(콜론, 들여쓰기)
* 들여쓰기는 2칸이면 2칸, 4칸이면 4칸으로 맞춰줘야 한다.
* if문이 끝나면 콜론을 꼭 써야한다.

###3. if문의 흐름
* 코드는 위에서 아래 순서대로 작동한다.

###4. if 중첩문
* 조건을 비교하고 맞지 않다면 다음 if문과 비교하여 출력함.

````python
# if문 문법과 주의사항(콜론, 들여쓰기)

개수 = int(input("붕어빵 몇개 사실껀가요?"))

if 개수 > 3:
    print("개당 1800원 입니다.")
````

````python
# if문의 흐름 (작동원리)
# 코드는 위에엇 아래 순서대로 작동한다.


# 2) 개수라는 변수에 담음 1) 입력을 받음
개수 = int(input("붕어빵 몇개 사실껀가요?"))

# 3) 개수 변수를 찾고, 입력된 값을 비교연산자를 통해 비교하여 True or False 값을 반환하여 조건에 맞게 실행함.
if 개수 > 3:
    print("개당 1800원 입니다.") # 조건이 True이면 if문 안의 코드가 실행됨


if 개수 <= 3:
    print("개당 2000원 입니다.")

# input의 값이 3을 초과한다면 개당 1800원, 3과 같거나 3보다 작다면 개당 2000원의 값을 출력할 것이다.
# if문 조건에 맞지 않다면 탈출하고 다른 조건과 비교하여 실행한다.
````

````python
# if 중첩문

의문의_숫자 = 10

if 의문의_숫자 > 5:
    if 의문의_숫자 < 15:
        print("의문의 숫자는 5보다 크고 15보다 작습니다.")
````

````python
# if ~ else 문법과 주의사항(콜론, 들여쓰기)
# True일때는 if문 안의 작성된 코드가 실행됨.
# False일때는 if문 안의 작성된 코드가 실행된다.

개수 = int(input("붕어빵 몇개 사실껀가요?"))


if 개수 > 3:
    print("개당 1800원 입니다.")
else: # else는 조건이 필요 없기 때문에 False값이 내려와도 밑의 코드를 실행하게 됨. ex) 끝에 있는 내용이 일괄적일 때 else를 사용함.
    print("개당 2000원 입니다.")
````

````python
# if ~ else문 흐름(작동원리)

개수 = int(input("붕어빵 몇개 사실껀가요?"))


if 개수 > 3: # 4라면 조건에 맞기 때문에 True를 출력후 바로 아래 조건을 실행
    print("개당 1800원 입니다.")
else: # False로 내려오면 상관없이 바로 아래 조건을 실행
    print("개당 2000원 입니다.")
````

````python
from re import T
# 조건문 조건(True, False)

# True
if True:
    print(True)
else:
    print(False)


# False
if False:
    print(True)
else:
    print(False)


# None
if None:
    print(True)
else:
    print(False)


# 0
if 0:
    print(True)
else:
    print(False)


#1
if 1:
    print(True)
else:
    print(False)


#1.1
if 1.1:
    print(True)
else:
    print(False)


#'hello'
if 'hello':
    print(True)
else:
    print(False) # 어떤 숫자나 문자열은 True로 출력함. 하지만 빈 칸이라면 False 출력함


#''
if '':
    print(True)
else:
    print(False)


#not
if not True:
    print(True)
else:
    print(False)


if not False:
    print(True)
else:
    print(False)
````

````python
# 정답

# 예제로 if문 익혀보기
# 문제 : 단어 a의 가운데 글자를 반환하는 코드
# 조건 : 단어의 길이가 짝수라면 가운데 두글자를 출력하면 됩니다.

# 입력값1 : "abcde" => c
# 입력값2 : "qwer" => we

# a = input()
# 코드를 작성해보자.

a = input()

가운데_인덱스 = int(len(a) / 2)

if len(a) % 2 != 0:
    print(a[가운데_인덱스])
else:
    print(a[가운데_인덱스-1 : 가운데_인덱스+1])
````
