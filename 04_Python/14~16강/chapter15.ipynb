# for문의 주의점

```python
# for문으로 만들어보기
# for 변수 in range(횟수) <- 숫자 범위만큼 연속적인 숫자를 만들어주는 range
#   print()
```

```python
# for문 만들기

for i in range(5):
    print("반복이 되네 신가하당!")
```

```python
# for문의 흐름 확인해보기

for i in range(5):
    print("신기행!", i) # for문의 흐름은 range부터 시작하여 i에 0~4까지의 값이 들어가고 for문을 만나 반복하게 됨.
```

```python
# 예제로 for문 익혀보기
# 문제 : 절대값 정수가 담긴 정수 변수와 그 정수의 부호를 차례 담은 부호 변수가 있다
# 두 변수를 이용해 실제 정수의 합을 구하시오. (True : 양수, False : 음수)

정수 = [10, 5, 8]
부호 = [True, False, True]

합 = 0

for k, v in zip(부호, 정수): # zip은 각 정수와 부호 변수를 순서대로 매칭시켜 정리해준다. ex> True:10 False:5 True:8
    if k:
        합 += v # k(부호)가 True값을 가진다면 총 합에 더한다. / 합은 0이기에 True일 시 v에 정수값을 합에 더하여 할당한다.
    else:
        합 -= v # k(부호)가 False값을 가진다면 총 합에서 뺀다. / else문이기에 이전 if문에서 내려온 False값을 무조건 실행한다.

print(f'총 합은 {합}입니다.')
```
