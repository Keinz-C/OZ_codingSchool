#### 함수와 매서드
> 함수 : 어디에 속해있지 않고 단독 모듈이라 함수를 그냥 호출하여 사용이 가능하다. ex : print(), input()  
> 매서드 : 함수의 클래스 안에 속해 있고 클래스의 멤버 변수들을 이용해서 구현된 것. ex : "햄스터".count("햄") / "햄스터".index("햄")

```python
"햄스터".count("햄")
```

```python
"햄스터".index("햄")
```

```python
# 형변환에는 뭐가 있었냐! 바로 int(), float(), str(), bool()
# 어...? 타입명을 그대로 쓰면 형변환을 시킬 수 있네!

정수 = input()
정수 = int(정수)
print(type(정수))

실수 = input()
실수 = float(실수)
print(type(실수))
```

```python
정수 = int(input())
print(정수, type(정수))

실수 = float(input())
print(실수, type(실수))
```

그 외 입력값의 형변환이 가능한 종류
* tuple()
* list()
* dict()
* set()

```python
넌센스, 정답 = input().split(",") # input은 하나의 값만 받음 / split 메서드는 앞의 함수와 관련된 모든 것을 출력함
print(f'넌센스 퀴즈 : {넌센스}')
print(f'정답 : {정답}')
```

```python
# 시퀀스 자료형에 다 지원되는 메서드

번호1, 번호2, 번호3, 번호4 = input().split(",")

번호1 = int(번호1)
번호2 = int(번호2)
번호3 = int(번호3)
번호4 = int(번호4)
print(type(번호1), 번호2, 번호3, 번호4)
```

```python
#map() / 내장함수에 대해서 공부하는 것이 좋음

번호1, 번호2, 번호3, 번호4 = map(int, input().split(","))
print(type(번호1), 번호2, 번호3, 번호4)
```

```python
# 레거시 코드 : 누군가 남겨두고 간 코드
# 리팩토링 : 결과의 변경 없이 코드의 구조를 재조정하는 것.
# input()으로 진행하는 리팩토링

# 팥붕 슈붕 잡붕 변수 지정
# boong__1 = "붕어빵(팥)"
# boong__2 = "붕어빵(슈크림)"
# boong__3 = "붕어빵(잡채)"

# input()을 활용한 리팩토리 진행
boong__1, boong__2, boong__3 = input('붕어빵(팥), 붕어빵(슈크림), 붕어빵(잡채)를 순서대로 개수를 입력하고 꼭 쉼표를 입력하고, 구매하지 않는 것은 0으로 표시.').split(',')

# 팥붕 슈붕 잡붕 가격
boong__1__1 = 2000
boong__2__1 = 2500
boong__3__1 = 3000

# 슈붕 잡붕 주문 갯수
order__boong2, order__boong3 = input('주문 갯수 표시 후 콤마만 사용해서 표시').split(",")

# 슈붕 잡붕 토탈 갯수와 가격
many__boong = order__boong2 + order__boong3
price__boong = (boong__2__1 * 2) + (boong__3__1 * 3)

# 최종 출력
print(f'주문이 완료되었습니다. {boong__2} {order__boong2}개, {boong__3} {order__boong3}개 총{many__boong}개 결제 금액은{price__boong}원 입니다.')
```

```python
# print() 함수의 기능을 알아보자!

print("초코칩", "민트", "샷추가", "휘핑크림")
print("초코칩", "민트", "샷추가", "휘핑크림", sep="/")
print("초코칩", "민트", "샷추가", "휘핑크림", sep=":")
print("초코칩", "민트", "샷추가", "휘핑크림", sep="   ")
print("초코칩", "민트", "샷추가", "휘핑크림", sep ='\n') # 윈도우는 ₩ 맥은 \이 역슬래쉬
```

```python
# end는 한 문장으로 표현함.
print("초코칩", "민트", "샷추가", "휘핑크림", end=" ")
print("초코칩", "민트", "샷추가", "휘핑크림", end=" ")
print("초코칩", "민트", "샷추가", "휘핑크림", end=" ")
```
