###list

* list란? 원소들이 연속적으로 저장되는 형태의 자료형을 말한다.
* 시퀀스 자료형과 같음
    * str(문자), list(), tuple, dict, set
* 만드는 방법은 어렵지 않음. 우선 대괄호[]로 묶어주고, 값들은 콤마로 구분한다.

```python
# 첫 list

리스트 = []
print(리스트)

리스트 = [23, "이거슨 리스트여!", 3.26, True, False]
print(리스트)
```

```python
# list 다시 한 번 선언!
리스트 = ["이", "렇", "게", "만", "들", "수", "있", "다"]
리스트
리스트2 = list() # int(), float()과 같이 형변환도 가능하다.
print(리스트2)

#range는 연속된 숫자를 생산하는 기능이 있다.
range(10)

# list(range(시작, 끝))을 출력함.
리스트_매직_스따또 = list(range(0,20))
리스트_매직_스따또
```

```python
# list(range(시작, 끝, 증가)) 슬라이싱과 비슷
리스트_매지꾸2 = list(range(0,20,2))
print(리스트_매지꾸2)

리스트_매지꾸3 = list(range(20,0,-1)) # 20개가 꽉 차면 나머지는 안보이기에 0까지 보일려면 21(index의 마지막 번호처럼)까지 작성해야함.
print(리스트_매지꾸3)

리스트_매지꾸4 = list(range(20,0))
print(리스트_매지꾸4)
```

```python
# 슬라이싱

로또 = [3, 5, 15, 33, 41, 44]

# 로또 변수에 있는 리스트의 인덱스 1번부터 2번까지의 값 출력
print(로또[1:3])

# 로또 변수의 마지막 위치 출력
print(로또[-1])

# 반대로 출력
print(로또[::-1])

# 로또 리스트안에 33이 있는지 확인
print(33 in 로또)

# 로또 리스트를 로또2 리스트와 합치기
로또2 = [2, 12, 15, 24, 33, 39]

로또합 = 로또 + 로또2

print(로또합)

# range를 이용해서 1부터 10사이에 짝수만 들어있는 짝수(변수명) 리스트를 만들기
짝수 = list(range(2, 12, 2)) # range함수는 range(start, stop, step)으로 start=시작 숫자부터 stop-1까지를 step의 간격으로 생성한다. 10을 생성하려면 11의 값을 입력하고, 해당 리스트 안에는 안보이지만 11까지 생성되어있음.
print(짝수) # 즉, 10까지 생성해라 = list(ragne(1, 11)) / 1~10 중에서 짝수만 골라라 list(range(2, 12, 2)) 1~10중에서 2부터 시작하여 10을 생성하기 위해 11까지 입력했기에 끝은 항상 +1의 값을 지님

# 짝수 리스트에 들어있는 값을 2배 늘리고 다시 짝수 변수에 담기
짝수 = 짝수 * 2
print(짝수)

# 짝수 리스트에 들어있는 요소의 갯수 구하기
print(f'{len(짝수)}개')

# 짝수 리스트에 3번째 인덱스 출력
print(짝수[3])

# len()함수를 이용해 인덱스 마지막 값을 출력
print(짝수[len(짝수)-1]) # len을 통하여 짝수의 총 요소를 구하고, -1을 하여 마지막 값을 출력. 값 출력을 위해 []을 사용.
```

```python
# list 값 넣고, 바꾸고, 지우고

mzfood = ["숙주", "분모자", "마라", "소세지", "소고기", "옥수수면"]

# 값을 넣는 방법 2가지
# 1. append() : 마지막 요소에 값을 넣음
# 2. insert() : 원하는 곳에 값을 넣음

mzfood.append("고수")
print(mzfood)

mzfood.insert(2, "탕후루")
print(mzfood)

# 값 지우기
del mzfood[-1]
print(mzfood)

del mzfood[0]
print(mzfood)

# 값 바꾸기
mzfood[-2] = "양고기"
print(mzfood)

# 문자열은 시퀀스라면 전부 가능한지 / 문자열은 시퀀스여도 할당할 수 없음
# 인사 = "반가워"
# 인사[2] = "웡"
```
