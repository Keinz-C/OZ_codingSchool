# dict(딕셔너리)
    * list, tuple, srt, range(범위안에 연속된 데이터를 만드는 시퀀스)
    * 공통적인 특징은, 데이터의 연관성이 없다.
    * dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재할때 사용.  
    (예시 : 게임 캐릭터 힘:30, 지능:20, 체력:40, 민첩:90)
    * dict는 중괄호{}

```python
# list[], tuple(), dict{}

햄찌맨 = {"힘":30, "지능":20, "체력":40, "민첩":90}
햄찌맨

# {키 : 값}
# {key1:value1, key2:balue2}

# 시퀀스 데이터는 인덱싱을 이용해 데이터를 찾는다.

# 햄찌맨 [1] dict는 인덱싱을 할때 키를 찾아줘야한다.
햄찌맨["민첩"]
print(f'햄찌맨님의 힘은 {햄찌맨["힘"]}입니다.')
```

```python
# 딕셔너리를 만드는 다양한 방법

# 빈 딕셔너리 만들기
딕셔너리 = {}
딕셔너리

딕셔너리2 = dict()
딕셔너리2

# 값을 넣어 만드는 방법
# dict()로 생성시 값 넣는 기호 "", : 주의하는 것이 좋다.
딕셔너리3 = dict(힘=30, 지능=20, 체력=40, 민첨=90)
딕셔너리3

# zip() : 병렬처리를 도와준다
딕셔너리4 = dict(zip(["힘", "지능", "체력", "민첨"], [30, 20, 40, 90]))
딕셔너리4
```

```python
# 값의 중복은 허락되는데, 키의 중복은 허락하지 않는다. - 키가 중복될 경우 가장 뒤에 입력된 키의 밸류값을 덮어씌원다. 단 밸류의 중복은 상관 없음.

햄찌맨 = {"힘":30, "지능":20, "체력":40, "민첩":90, "힘":40, "지구력":90}
햄찌맨

"체력" in 햄찌맨
```

* 해시, 해시맵, 해시테이블과 관련이 높음.(자료구조 / 키와 밸류에 관련된 것이 해시)

```python
# 팥붕 슈붕 잡붕 변수 지정
boong__1 = "붕어빵(팥)"
boong__2 = "붕어빵(슈크림)"
boong__3 = "붕어빵(잡채)"

# 팥붕 슈붕 잡붕 가격
boong__1__1 = 2000
boong__2__1 = 2500
boong__3__1 = 3000

# 슈붕 잡붕 주문 갯수
order__boong2 = 2
order__boong3 = 3

# 슈붕 잡붕 토탈 갯수와 가격
many__boong = order__boong2 + order__boong3
price__boong = (boong__2__1 * 2) + (boong__3__1 * 3)

# 최종 출력
print(f'주문이 완료되었습니다. {boong__2} {order__boong2}개, {boong__3} {order__boong3}개 총{many__boong}개 결제 금액은{price__boong}원 입니다.')
```

```python
주문_갯수 = map(int, input(f'붕어빵 맛 순서대로 갯수를 입력 안살꺼면 0입력 구분은 콤마(팥, 슈크림, 잡채)').split(','))

품목별_가격 = {"팥":2000, "슈크림":2500, "잡채":3000}

주문_갯수 = list(주문_갯수) # 딕셔너리는 인덱싱이 불가하기 때문에 list를 사용하여 인덱싱이 가능하도록 변경한다.
총_주문_갯수 = sum(주문_갯수) # sum은 list의 모든 값을 더해준다.
total = 0 # 변수을 초기화하기 위해 0을 대입한다.

total += 품목별_가격["팥"] * 주문_갯수[0]
total += 품목별_가격["슈크림"] * 주문_갯수[1]
total += 품목별_가격["잡채"] * 주문_갯수[2]

print(f'주문이 완료되었습니다. 팥{주문_갯수[0]}개, 슈크림{주문_갯수[1]}개, 잡채{주문_갯수[2]}개, 총{총_주문_갯수}개 결제 금액은{total}원 입니다.')
```
