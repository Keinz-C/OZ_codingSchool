* set
        * set은 수학의 집합을 의미한다.
        * 딕셔너리와 같이 순서가 없음.
        * 값의 중복을 허락하지 않는다. *중요*
        * set의 생성 방법은 중괄호

```python
채소 = {"당근", "배추", "대파", "양파"}
type(채소)
print(채소)

# 순서가 없고, 출력하면 요소가 다르게 나오는 것도 특징

# set를 만드는 다양한 방법
세트 = set()
세트
세트2 = {}
세트2
```

```python
# 중복을 제일 싫어하는 set

바나나_리스트 = list("banana")
print(바나나_리스트)
바나나_튜플 = tuple("banana")
print(바나나_튜플)
바나나_셋 = set("banana")
print(바나나_셋)
```

```python
# set() 자료형의 데이터 추가, 삭제, 변경
채소

# set() 함수의 데이터 삽입 / add는 데이터 하나가 온전한 하나의 요소로 인정, update는 데이터의 각 글자 하나하나 개별요소로 인식하여 삽입하여 오류가 생길 수 있음.
채소.add("토마토")
채소

채소.update("깻잎")
채소

# set() 함수의 데이터 삭제
# 채소.remove("잎")
# 채소

# 채소.remove("잎")
채소.discard("잎")
채소.discard("깻")
채소.discard("조")
채소.discard("림")
채소

# 임의의 데이터를 삭제하고 반환
채소.pop()
채소
```

```python
채소 = {"당근", "배추", "대파", "양파"}
반환된_채소 = 채소.pop()

print(f'{반환된_채소}는 내다버린 채소입니다.')
```

```python
# set = 집합, 합집합, 교집합, 차집합 기타 등등!
# 대표적으로 합집합, 교집합, 차집합

채소1 = {"당근", "배추", "대파", "양파"}
채소2 = {"브로콜리", "양파", "쪽파", "오이"}

# 합집합(union)
채소 | 채소2
print(f'{set.union(채소1, 채소2)}합집합 입니다.')

# 교집합(intersection)
채소1 & 채소2
print(f'{set.intersection(채소1, 채소2)}교집합 입니다.')

# 차집합(difference)
채소1 - 채소2
print(f'{set.difference(채소1, 채소2)}차집합 입니다.')
```
