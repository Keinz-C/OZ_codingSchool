이러한 구조를 바탕으로 oz()함수를 만들어보자
####요구조건
- start, end, multiple 3개의 매개변수를 이용해 start부터 시작해 multiple씩 증가해 end에 도달하면 종료하는 함수  
예) start = 1, end = 10, multiple = 3 (출력) => 1,4,7

````python
# start, end, multiple 3개의 매개변수를 이용해 start부터 시작해 multiple씩 증가해 end에 도달하면 종료하는 함수
# 예) start = 1, end = 10, multiple = 3 (출력) => 1,4,7

def oz(start, end, multiple):
    for i in range(start, end, multiple):
        print(i, ",", end="")
    print()

oz(1, 10, 3)
oz(start=1, end=20, multiple=4)     # 코드가 길어질 경우 시인성이 좋은건 키워드 매개변수를 사용하는 것이 좋다.
oz(start=1, end=30, multiple=5)
````

````python
def oz(start=1, end=12, multiple=1):    # 기본 매개변수로 변경하여도 함수를 호출할 때 인수 값을 지정하면 된다.
    for i in range(start, end, multiple):
        print(i, ",", end="")
    print()

oz(1, 10, 3)
oz(multiple=4)
oz(end=30, multiple=10)
````
