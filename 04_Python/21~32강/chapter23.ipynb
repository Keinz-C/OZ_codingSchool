```
마치 값을 할당하는 형태처럼 보이는 이 매개변수들은 모두 기본 매개변수이다.
```
기본 매개변수는 매개변수를 입력하지 않으면 함수 정의시 설정해 놓은 기본값으로 대체된다.  
역시나, 기본 매개변수 사용시에도 제약은 있다.
- 기본 매개변수 뒤에는 일반 매개변수가 올 수 없다.

````python
def oz(value, n=2):     # 여기서 value는 일반 매개변수, n=2는 기본 매개변수에 속한다. n=2에서 뒤에 2로 값을 할당할 때 이를 기본 매개변수라고 한다.
    for i in range(n):
        print(value)

oz("일반 매개변수 다음에 기본 매개변수를 배치해야 한다.")
````

````python
# 기본 매개변수의 값을 변경할 수 있다.
def oz(value, ddddd):
    for i in range(ddddd):
        print(value)

oz("일반 매개변수 다음에 기본 매개변수를 배치해야 한다.", ddddd=10)
````

````python
# 기본 매개변수를 일반 매개변수보다 먼저 썼을 때
def oz(n=2, value):
    for i in range(n):
        print(value)

oz(n=10, "일반 매개변수 다음에 기본 매개변수를 배치해야 한다.")
````
