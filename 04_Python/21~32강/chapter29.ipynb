튜플은 list와 비슷한 시퀀스 자료형이다. 단, list와의 차이점은 요소를 바꿀 수 없다는 점이다.  

튜플을 생성할때는 소괄호 "()"을 이용한다.  

변수명 = (데이터1, 데이터2, 데이터3)

튜플을 이용해 괄호 없이 여러 값을 할당할 수 있다.

```python
oz_tuple = 1, 2, 3, 4, 5, 6

print(oz_tuple)
print(type(oz_tuple))
```

```python
x, y, z = 1, 2, 3
print(x)
print(y)
print(z)
print(type(x))
print(type(y))
print(type(z))
```

튜플 내부의 값을 변경할 수는 없지만, 교환할 수는 있다.

```python
x, y = 1, 2

print("x:", x)
print("y:", y)
print()

x, y = y, x
print("x:", x)
print("y:", y)
```

튜플을 이용하면 여러 개의 값을 리턴할 수 있다.

```python
def oz_tuple():
    return(1, 2)

x, y = oz_tuple()
print(x)
print(y)
```

```python
# 튜플을 이용한 내부함수들
# enumerate()의 i, fruit는 (i, fruit)이나 괄호를 생략한 상태이다.

for i, fruit in enumerate([1, 2, 3, 4]):
    print(i, fruit)
```

```python
# divmode() 사용 시 튜플 형태로 몫과 나머지를 반환

a, b = 7, 3
x, y = divmod(a, b)
print("몫:", x)
print("나머지:", y)
```
