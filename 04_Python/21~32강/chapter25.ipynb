### return 키워드
- 함수를 실행한 위치로 되돌아 감, 일반적으로 함수가 끝나는 위치에 return을 사용함

```python
# 값 없이 리턴하는 경우

def oz_return():
    print("안녕")
    print("파이썬")
    return
    print("개꿀잼아니죠~")

oz_return()
```

1. oz_return()로 함수를 호출하면 함수안 코드가 위엥서부터 하나씩 실행됨
2. 세번째에서 return을 만나 oz_return()으로 함수를 호출했던 그 위치로 돌아감

```python
# 값과 함께 리턴하는 경우

def oz_return2():
    return "파이썬 재밌엉"  # return옆은 단순히 문자열이기 때문에 반환된다고 화면에 출력되지 않는다.

result1 = oz_return2()   # return에서 반환된 값을 출력하기 위해 변수에 담고 출력하면 된다.
print(result1)
```

```python
# 아무것도 리턴하지 않는 경우

def oz_return3():
    return

result = oz_return3()
print(result)
```

```python
# bool 함수를 통해 위의 None에 대한 결과값이 True인지 False인지 확인하기

def oz_return2():
    return "파이썬 재밌엉"

result1 = oz_return2()
print(result1)
print(bool(result1))    # 이처럼 값을 반환받아 출력되는 경우 bool함수를 통해 확인하면 True값이 출력된다.

def oz_return3():
    return

result3 = oz_return3()
print(result3)
print(bool(result3))    # 이처럼 return을 통한 반환값이 없을 경우 None을 띄우며 그 값은 False가 출력된다.
```
