```python
#(기본 매개변수, 가변 매개변수)

def oz(n=3, *values):
    for i in range(n):
        for value in values:
            print(value)
            print()

oz("기본 매개변수 다음에", "가변 매개변수를", "넣을 수 없어!")
```

위를 해석하면  
기본 매개변수인 n에는 "기본 매개변수 다음에"라는 값이 들어가고  
가변 매개변수에는 "가변 매개변수를", "넣을 수 없어!"가 들어가야 하지만,  
앞의 기본 매개변수에서 오류가 걸렸기 때문에 들어가지 않았다.  

이 때문에 기본 매개변수가 가변 매개변수보다 앞에 오면 기본 매개변수의 의미가 사라진다.

```python
# 가변 매개변수 뒤 기본 매개변수가 오는 코드
def oz(*ff, n=3):
    for i in range(n):
        for value in ff:
            print(value)
            print()

oz("기본 매개변수 다음에", "가변 매개변수를", "넣을 수 없어!", n=2)  # 숫자만 넣을시 기본 매개변수를 지정하고 값을 넣어야한다. ex) 10 (x) / n = 10(o)
```

기본 매개변수에 10을 넣으려 했으나, 이를 가변 매개변수로 인식하여  
기본 매개변수는 함수의 기본값인 3번이 반복되었다.
이러한 문제를 해결하기 위해서 나온 개념이 <span style="color: red">키워드 매개변수</span>이다.

앞서 몇번 언급했던 print()함수를 다시 한 번 보면 기본 매개변수가 여럿 있다.
그 중에서도 end는 꽤나 익숙할 수 있다.

```python
# end = \n
list = ["안","녕","하","세","요"]
for i in list:
    print(i)
```

```python
# end 키워드 매개변수를 이용한 한 줄 출력
list = ["안","녕","하","세","요"]
for i in list:
    print(i, end="")
```

print() 함수를 호출할 때 i는 가변 매개변수로 입력되고 뒤에 나오는 end="" 기본 매개변수가 된다.  
앞선 예시 코드와 다른 점은 end=""라는 형식으로 값을 지정한 부분이다.  
end=""이 부분이 오늘 수업의 핵심인 <span style="color:red">키워드 매개변수</span>이다.

```python
# 직육면체를 부피를 구하는 함수 : 가로 = x, 세로 = y, 높이 = h

def oz(x, y=20, h=10):
    print(x * y * h)

# x식별자에 매개변수 10을 넣은 경우

oz(10)

# x식별자에 10을 넣고 y=10이라는 키워드 매개변수를 이용한 경우

oz(10, y = 10)

# 모든 매개변수를 키워드 매개변수로 넣은 경우

oz(x=1, y=200, h=105)
```

키워드 매개변수는 이렇게 다양하게 사용될 수 있다.
