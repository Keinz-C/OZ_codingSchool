###제너레이터(generator)
1. 이터레이터를 생성하는 객체.  
(이터레이터 : 순서대로 다음 값을 리턴할 수 있는 객체로 자체적으로 내장하고 있는 next메소드를 통해 다음 값을 가져올 수 있다.)  

제네레이터는 함수안에 yield 키워드를 사용해 만들 수 있다.  

특이하게 함수 안에 yield라는 키워드가 어떤 동작을 하든 하지 않든 있기만 해도 제너레이터로 인식된다.  

단, 제너레이터 함수가 되면 호출해도 함수 내부의 코드가 작동하지 않는다.

```python
def test():
    print("제너레이터 실습 코드입니다.")
    yield "test"        # generator함수가 되었기 때문에 함수 내부의 프린트문은 출력되지 않는다.

print("첫번째 출력")
test()

print("두번째 출력")
test()

print(test())
```

```python
# generator함수를 실행시키기 위한 코드

def test():
    print("첫번째 출력")
    yield 1
    print("두번째 출력")
    yield 2
    print("세번째 출력")

output = test()
print("네번째 출력")

x = next(output)        # test 제너레이터 함수를 output에 담고, output을 next함수의 매개변수로 담은 것이다.
print(x)
print("다섯번째 출력")
y = next(output)
print(y)
print("여섯번째 출력")
z = next(output)        # yield가 없는 상태에서 next함수를 사용했기 때문에 오류가 발생한다.
print(z)
```

위 실행 결과처럼 next()함수를 만나면 yield 키워드를 만날때까지 실행된다.  
첫번째 yield 1을 만나기 전까지 첫번째 출력이라는 print문이 작동하고, yield의 값 1이 리턴된다.  

이후 세번째 next(output)이 실행되는데 test() 함수안에 yield가 2개 밖에 없기 때문에 세번째 실행 시 StopIteration이라는 예외가 발생한다.  

###generator 함수를 사용하는 상황
- 코드를 조금 실행하는 상황에서 자주 사용된다.
    - 코드를 조금씩 실행시키게 되면, 함수 전체를 실행하지 않아도 되기 때문에 메모리를 효율적으로 쓸 수 있다.
