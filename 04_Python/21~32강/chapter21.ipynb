함수의 구조
> def 함수이름(매개변수, 매개변수):  
&emsp;코드

함수의 괄호안에 들어오는 값을 매개변수(parameter)라고 한다.  
이렇게 입력되는 매개변수들을 구분하기 위해 식별자를 만든다.

```python
def oz(message, n):
    for i in range(n):
        print(message)

oz('oz코딩스쿨', 5)
```

위 코드는 앞서 만든 oz()함수에 2개의 매개변수를 받을 수 있도록 수정  
2개의 매개변수는 각각 message와 n이라는 식별자를 만듦.  

이로써 oz()함수 호출시 입력하는 2개의 값 중 첫 번째는 출력에 사용되는 값이고 두 번째는 for문 반복에 사용됨.  
구조의 큰 변화는 없지만 이전 oz()함수와는 작동 방식에 차이가 있음.
