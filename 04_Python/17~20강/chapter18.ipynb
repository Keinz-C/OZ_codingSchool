###self란?
1. 인스턴스 자기 자신을 의미한다.
2. 여기서 __init__의 매개변수 self에 들어가는 값은
Person()이라고 할 수 있음

class Person:  
&emsp;def __init__(self):  
&emsp;&emsp;self.hello = '안녕하세요'

&emsp;def greeting(self):  
&emsp;&emsp;print(self.hello)


**순서**

1. class 함수에서 Person이라는 클래스 이름 선언(클래스 이름은 대문자로 시작해야 한다.)
2. 메서드를 작성할 때, 메서드의 첫 번째 매개변수는 반드시 self로 지정해야한다.(__init__또한 기본 구성임.)
3. 기능시킬 코드를 입력

```python
class 클래스이름:
    def __init__(self, 매개변수1, 매개변수2):
        self.속성1 = 매개변수1
        self.속성2 = 매개변수2
```

```python
class Person:
    def __init__(self, name, age, address):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.address = address

    def greeting(self):
        print('{0} 저는 {1}입니다.'.format(self.hello, self.name)) #<- maria.name으로 바꿔도 출력되는건 아래 maria.greeting으로 프린트문을 출력하는 것임.
                                                                # class 안에 maria가 없기 때문에 애초에 값이 출력되지 않는것이 맞음

maria = Person('마리아', 20, '서울시 서초구 반포동')
maria.greeting()# 안녕하세요. 저는 마리아입니다.

print('이름:', maria.name)      # 마리아
print('나이:', maria.age)       # 20
print('주소:', maria.address)   # 서울시 서초구 반포동


# 순서를 적어보자
# 1. maria라는 인스턴스를 통해 Person이라는 클래스가 호출.
# 2. __init__가 초기화를 시킴. 각 매개변수가 ()안의 값을 self(초기화값)을 제외한 매개변수에 할당.
# 3. 두번째 함수를 통해 self를 호출하여, 프린트문을 출력하게 함.
# 3-1. 여기서 .format은 f-string과는 사뭇 다르게 ()안의 값을 순서대로 할당하고 출력시킴.
# 4. 마지막의 각 프린트문이 maria라는 인스턴스와 속성들을 통해 값을 불러 출력함.
```

## 클래스의 상속

- 기반 클래스(base class) : 기능을 물려줄 클래스
- 파생 클래스(derived class) : 기능을 물려받은 클래스

```python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print('멈췄습니다.')

class HybridCar(Car):
    battery = 1000
    batteryKM = 300

k5 = HybridCar()
k5.maxspeed
print(k5.maxspeed)

print(dir(k5))

# 분석
# 1. k5라는 인스턴스가 HybridCar를 호출
# 2. class HybridCar(Car)는 기반 클래스인 Car의 기능을 물려받음.
# 3. class Car는 매개변수 object를 가지고, 여기에 maxspeed와 maxpeople이라는 추가적인 매개변수를 가지고 있음
# 4. k5.maxspeed를 불러올 때, 본래 HybridCar에는 없는 매개변수지만, 기반 클래스인 Car의 기능을 받았기 때문에 출력 가능.
```

```python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print('멈췄습니다.')

class HybridCar(Car):
    battery = 1000
    batteryKM = 300

class ElectricCar(HybridCar):
    battery = 1000
    batteryKM = 300

k5 = HybridCar()
# k5.maxspeed

electricCark5 = ElectricCar()
electricCark5.maxspeed

# 분석
# 결과 : 300
# 1. 마지막 electricCark5라는 인스턴스가 ElectricCar를 호출한다.
# 2. 파생클래스 ElectricCar는 HybridCar의 매개변수를, HybridCar는 Car의 매개변수 값을 가진다.
# 3. ElectricCar 클래스에는 Car의 매개변수가 있기에 결과값인 maxspeed의 값을 출력할 수 있다.
```

```python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print('멈췄습니다.')

class HybridCar(Car):
    battery = 1000
    batteryKM = 300

class ElectricCar(HybridCar):
    battery = 1000
    batteryKM = 300
    def move(self, x):
        print(self.batteryKM, '만큼 달릴 수 있습니다.')
        print(x, '스피드로 달리고 있습니다.')

k5 = HybridCar()
k5.maxspeed

electricCark5 = ElectricCar()
electricCark5.maxspeed
electricCark5.battery
electricCark5.move(10)

# 분석
# 결과 : 300만큼 달릴 수 있습니다. / 10 스피드로 달리고 있습니다.
# 1. 각 파생 클래스는 기반 클래스의 값을 가진다.
# 2. 인스턴스로 move에 10값을 넣어 호출하기 때문에 move의 매개변수 x는 10의 값을 가진다.
# 추가 : 메서드 오버라이딩으로 파이썬에서는 맨 마지막에 있는 메서드만 인정을 하기 때문에, ElectricCar class의 프린트문이 출력된다.
```
