```python
f = open('text.txt','w')
message = ''

for i in range(10):
    message += "개발자 가즈아!!!\n"
f.write(message)
f.close()
```

```python
f = open('text.txt','r')
s = f.read() # 읽어온 내용은 저장을 해야지 다음 프린트문에서 읽어올 수 있음.
print(s)
f.close()
```

```python
f = open('text.txt','r')
line = f.readline()     # readline은 저장된 문서에서 한 줄 씩 읽어오는 것이다.
print(line)
print(type(line))
f.close()
```

```python
list = [10, 20, 30]

with open('text.txt', 'a') as f:
    f.write(str(list))
# 추가 작성등은 문자형으로밖에 안되기에 list를 추가하려면 형변환이 필요함.
```

```python
list = [10, 20, 30]
with open('text.txt', 'a') as f:
    f.write('let test =' + str(list))
    # let은 중복을 허용하지 않는다.
    # list에는 배열될 값을 넣음
```

```python
list = ['10', '20', '30']

with open('text.txt', 'a') as f:
    f.writelines(list)      # writelines는 ()안의 값을 하나하나 읽어내기 때문에 각각 문자화 시켜줘야한다
```

```python
#csv

list = ['10', '20', '30']       # 문자열밖에 받지 못하기 때문에 전부 문자화시켜줘야함.

with open('text3.csv', 'w') as f:
    f.write(','.join(list))     # 데이터를 구분하는 기준은 다양함
```
