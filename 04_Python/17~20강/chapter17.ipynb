```python
# global 전역변수는 함수 안으로 호출 할 수 없기 때문에 억지로 지정하여 가져오는 수밖에 없다.
a = 100
def f():
    global a
    a = a + 1
    return a
f()
```

```python
# dict 타입에 한해서 전역변수 값을 지역변수에 호출하여 사용할 수 있다.
won = {'A등급':1000, 'B등급':500, 'C등급':100}

def 계산(a, b, c):
    합계 = won['A등급']*a + won['B등급']*b + won['C등급']*c
    return 합계

print(계산(5, 2, 3), '원입니다.')
```

```python
#상수
# 바깥에 있는 전역변수를 아규먼트에 할당하고, 계산하면 매개변수가 계산을 할 수 있음.
PI = 3.14
def circle(r, inputpi):
    z = r * r * inputpi
    return z

result = circle(10, PI) # 아규먼트
print(result)
```

```python
A = 0
B = 0
C = 0
유물 = 0

def 감자():
    global A
    A += 1

def 양파():
    global A, B, C
    A += 3
    B += 3
    C += 4

def 옥수수():
    global 유물
    유물 += 1

감자()
양파()
옥수수()

합계 = A + B + C + 유물
print('오늘 수확한 수확물 :', 합계)
```

```python
def 수확철():
    야채 = {'A등급':0, 'B등급':0, 'C등급':0}

    def 양파():
        야채['A등급'] += 1

    def 대파():
        야채['A등급'] += 3
        야채['B등급'] += 3
        야채['C등급'] += 4

    def 쪽파():
        야채['금'] = 1

    양파()
    대파()
    쪽파()

    return 야채

수확철()
# sum(수확철().values())
# 연산자 우선순위를 계속 확인하는 것이 좋다.
```
