###예외처리

- 크롤링이나 서비스에 문제가 발생할 경우 어떻게 처리, 대처해야 될 때 사용하는 문법
- 무중단 서비스를 운영하는 경우 중요한 문법이 됨.

```python
try:
    x = int(input('나눌 숫자를 입력하세요'))
    y = 10 / x
    print(y)
except:
    print("예외가 발생했습니다.")
```

```python
y = [10, 20, 30]
try:
    index, x = map(int, input('나눌 숫자를 입력하세요:').split(',')) # index는 처음 입력하는 정수가 할당되고 콤마로 구분하여 다음 오는 숫자를 x에 할당한다.
    print(y[index]/x) # y라고 변수명을 잡은 리스트 안에서 인덱스 값을 추출하여 x로 나눈다.
except:
    print('오류입니다.')
```

```python
# ZeroDivisionError의 경우 숫자를 0으로 나눴을 때 에러를 처리하는 기능이다.
y = [10, 20, 30]
try:
    index, x = map(int, input('나눌 숫자를 입력하세요:').split(','))
    print(y[index]/x)
except ZeroDivisionError:
    print('숫자를 0으로 나눌 수 없습니다.')
except:
    print('에러에러')
```

```python
# IndexError의 경우 잘못된 인덱스값을 입력할 때 나타나는 에러를 처리하는 기능이다.
c
```

```python
try:
    x = int(input('나눌 숫자를 입력하세요'))
    y = 10 / x
    print(y)
except:
    print("예외가 발생했습니다.")
else:
    print('출력에 성공하였습니다.')        # else : 정상적으로 실행됐을 때, 실행되는 코드를 작성하는 공간이다.
```

```python
try:
    x = int(input('나눌 숫자를 입력하세요'))
    y = 10 / x
    print(y)
except:
    print("예외가 발생했습니다.")
else:
    print('출력에 성공하였습니다.')
finally:
    print('코드 실행이 종료되었습니다.')    # finally : 위의 여부 상관 없이 무조건 실행하는 코드이다.
```

```python
# raise : 임의로 Error를 만드는 방법
try:
    x = int(input('나눌 숫자를 입력하세요'))
    y = 10 / x
    raise
    print(y)
except:
    print("예외가 발생했습니다.")
else:
    print('출력에 성공하였습니다.')
finally:
    print('코드 실행이 종료되었습니다.')
```
