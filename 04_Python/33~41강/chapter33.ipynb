###오류의 종류
프로그램 언어의 오류에는 두 가지 종류가 있다.
- 프로그램 실행 전에 발생하는 오류를 구문 오류라고 함
- 프로그램 실행 중에 발생하는 오류를 예외 또는 런타임 오류라고 함  

###[구문오류]
- 괄호의 개수, 들여쓰기 문제 등으로 프로그램이 실행되기도 전에 발생하는 오류

```python
# 구문오류의 예제

print("오류는 사소한 것부터 체크하자")
```

```python
# 닫는 괄호를 생략한 경우
print("오류는 사소한 것부터 체크하자"
```

**SyntaxError** : 구문에 문제가 발생해 프로그램이 실행조차 되지 않는 오류  
-> 이런 경우 구문 오류를 해결하지 않으면 프로그램을 실행할 수 없음

```python
# 생략한 괄호를 삽입해 구문오류 해결
print("오류는 사소한 것부터 체크하자")
```

###[기본 예외 처리]
예외를 해결하는 모든 것을 예외 처리라고 함  

예외 처리하는 2가지 방법
- 조건문을 사용하는 방법
- try 구문을 사용하는 방법  

단, 구문 오류가 발생하는 경우는 예외 처리로 불가하기 때문에 코드 자체를 수정해야 함.

[조건문을 이용한 예외 처리]
- 조건문을 이용한 예외 처리방법을 기본 예외 처리라고 함.  

예외가 발생할 수 있는 상황을 파악하는 게 우선

```python
x, y = map(int, input("밑변과 높이를 정수로만 입력해주세요.").split(" "))


print('삼각형 넓이는 {0}입니다.'.format(x*y/2))
```

```python
# x = 3cm, y = 5cm
x, y = map(int, input("밑변과 높이를 정수로만 입력해주세요.").split(" "))


print('삼각형 넓이는 {0}입니다.'.format(x*y/2))
```

정수를 입력하지 않으면 위 코드는 ValueError가 발생한다.  

조건문에 사용할 함수는 isdigit()이다.
- isdigit() : 숫자로만 구성되어있는 글자인지를 판단하고 숫자로 변환이 가능한 글자라면 True를 반환한다.
  
예) num = "3"  
isdigit(num) => True

```python
x, y = input("밑변과 높이를 정수로만 입력해주세요.").split(" ")

if x.isdigit() and y.isdigit():
    x, y = int(x), int(y)
    print('삼각형 넓이는 {0}입니다.'.format(x*y/2))
else:
    print('정수로 입력하지 않아 계산이 불가합니다.')
```
